Visualizing The Gender Gap In College Degrees 

In this project, I aim to visualize the gender gap in college degrees. The US Department of Education released a dataset containing the percentage of bachelor's degrees granted to women from 1970 to 2012. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('H:/Dataquest/Data Analyst/Step2/Datasets/percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

Defining the degrees into 3 groups

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(18, 18))

## STEM Degree charts.
for sp in range(0,18,3):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[int(sp/3)]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[int(sp/3)]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[int(sp/3)])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if int(sp/3) == 0:
        ax.text(2005, 85, 'Men')
        ax.text(2002, 10, 'Women')
    elif int(sp/3) == 5:
        ax.text(2005, 89, 'Men')
        ax.text(2001, 8, 'Women')
        
## Liberal Arts Degree charts.        
for sp in range(1,16,3):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[int((sp-1)/3)]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[int((sp-1)/3)]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[int((sp-1)/3)])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if int((sp-1)/3) == 0:
        ax.text(2005, 78, 'Men')
        ax.text(2002, 18, 'Women')
        
## Other Degree charts.        
for sp in range(2,20,3):
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[int((sp-2)/3)]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[int((sp-2)/3)]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[int((sp-2)/3)])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if int((sp-2)/3) == 0:
        ax.text(2005, 91, 'Men')
        ax.text(2002, 5, 'Women')
    elif int((sp-2)/3) == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 30, 'Women')
        
plt.show()
```

We should keep the data:ink ratio as high as possible. Which means that the ink used to show data should be high and the 'non-data ink' should be minimum.
